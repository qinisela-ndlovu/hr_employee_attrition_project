This Notebook covers Data Cleaning, for a bigger project

```python
#import the required libraries for data manipulation ..
import pandas as pd
HR_data = pd.read_csv('Data Science source data/IBM-HR Data Analytics-Employee-Attrition.csv')
```

```python
#Show first few rows of the dataset
HR_data.head()
```

```python
#Basic exploration
row,col = HR_data.shape
print('There are', str(row) +' rows and', str(col) +' columns in this dataset')
HR_data.info()
```

The are no missing values in the dataset

For better analysis and memory, deal with categorical values(columns)


```python
# Identify columns with object data type
categorical_cols = HR_data.select_dtypes(include=['object']).columns

# Convert each of these columns to 'category' type
for col in categorical_cols:
    HR_data[col] = HR_data[col].astype('category')

# Print the names of all categorical columns
print("Categorical columns:")
print(categorical_cols)
```

```python
#check for duplicates
HR_data.duplicated().sum()
```

```python
# Add an a new column to convert the categorical "Attrition" column (Yes / No) into a numeric value (1 / 0) 
HR_data['Attrition_Value'] = HR_data['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
```
